```python
import requests
import zipfile
import os
import cv2
import numpy as np,pandas as pd
import glob
from sklearn.model_selection import train_test_split
from tensorflow.keras.preprocessing.image import load_img, img_to_array
import tensorflow as tf
from tensorflow.keras import layers, models

# Step 1: Download the dataset into a 'data' folder
urls = {
    "fall-01-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-01-cam0-d.zip",
    "fall-01-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-01-cam0-rgb.zip",
    "fall-01-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/fall-01-data.csv",
    "fall-02-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-02-cam0-d.zip",
    "fall-02-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-02-cam0-rgb.zip",
    "fall-02-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/fall-02-data.csv",
    "fall-03-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-03-cam0-d.zip",
    "fall-03-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-03-cam0-rgb.zip",
    "fall-03-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/fall-03-data.csv",
    "fall-04-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-04-cam0-d.zip",
    "fall-04-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-04-cam0-rgb.zip",
    "fall-04-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/fall-04-data.csv",
    "fall-05-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-05-cam0-d.zip",
    "fall-05-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-05-cam0-rgb.zip",
    "fall-05-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/fall-05-data.csv",
    "fall-06-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-06-cam0-d.zip",
    "fall-06-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-06-cam0-rgb.zip",
    "fall-06-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/fall-06-data.csv",
    "fall-07-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-07-cam0-d.zip",
    "fall-07-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-07-cam0-rgb.zip",
    "fall-07-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/fall-07-data.csv",
    "fall-08-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-08-cam0-d.zip",
    "fall-08-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-08-cam0-rgb.zip",
    "fall-08-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/fall-08-data.csv",
    "fall-09-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-09-cam0-d.zip",
    "fall-09-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-09-cam0-rgb.zip",
    "fall-09-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/fall-09-data.csv",
    "fall-10-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-10-cam0-d.zip",
    "fall-10-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-10-cam0-rgb.zip",
    "fall-10-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/fall-10-data.csv",
    "fall-11-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-11-cam0-d.zip",
    "fall-11-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-11-cam0-rgb.zip",
    "fall-11-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/fall-11-data.csv",
    "fall-12-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-12-cam0-d.zip",
    "fall-12-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-12-cam0-rgb.zip",
    "fall-12-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/fall-12-data.csv",
    "fall-13-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-13-cam0-d.zip",
    "fall-13-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-13-cam0-rgb.zip",
    "fall-13-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/fall-13-data.csv",
    "fall-14-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-14-cam0-d.zip",
    "fall-14-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-14-cam0-rgb.zip",
    "fall-14-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/fall-14-data.csv",
    "fall-15-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-15-cam0-d.zip",
    "fall-15-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-15-cam0-rgb.zip",
    "fall-15-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/fall-15-data.csv",
    "fall-16-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-16-cam0-d.zip",
    "fall-16-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-16-cam0-rgb.zip",
    "fall-16-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/fall-16-data.csv",
    "fall-17-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-17-cam0-d.zip",
    "fall-17-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-17-cam0-rgb.zip",
    "fall-17-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/fall-17-data.csv",
    "fall-18-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-18-cam0-d.zip",
    "fall-18-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-18-cam0-rgb.zip",
    "fall-18-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/fall-18-data.csv",
    "fall-19-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-19-cam0-d.zip",
    "fall-19-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-19-cam0-rgb.zip",
    "fall-19-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/fall-19-data.csv",
    "fall-20-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-20-cam0-d.zip",
    "fall-20-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/fall-20-cam0-rgb.zip",
    "fall-20-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/fall-20-data.csv",

    "adl-01-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/adl-01-cam0-d.zip",
    "adl-01-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/adl-01-cam0-rgb.zip",
    "adl-01-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/adl-01-data.csv",
    "adl-02-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/adl-02-cam0-d.zip",
    "adl-02-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/adl-02-cam0-rgb.zip",
    "adl-02-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/adl-02-data.csv",
    "adl-03-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/adl-03-cam0-d.zip",
    "adl-03-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/adl-03-cam0-rgb.zip",
    "adl-03-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/adl-03-data.csv",
    "adl-04-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/adl-04-cam0-d.zip",
    "adl-04-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/adl-04-cam0-rgb.zip",
    "adl-04-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/adl-04-data.csv",
    "adl-05-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/adl-05-cam0-d.zip",
    "adl-05-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/adl-05-cam0-rgb.zip",
    "adl-05-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/adl-05-data.csv",
    "adl-06-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/adl-06-cam0-d.zip",
    "adl-06-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/adl-06-cam0-rgb.zip",
    "adl-06-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/adl-06-data.csv",
    "adl-07-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/adl-07-cam0-d.zip",
    "adl-07-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/adl-07-cam0-rgb.zip",
    "adl-07-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/adl-07-data.csv",
    "adl-08-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/adl-08-cam0-d.zip",
    "adl-08-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/adl-08-cam0-rgb.zip",
    "adl-08-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/adl-08-data.csv",
    "adl-09-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/adl-09-cam0-d.zip",
    "adl-09-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/adl-09-cam0-rgb.zip",
    "adl-09-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/adl-09-data.csv",
    "adl-10-cam0-d.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/adl-10-cam0-d.zip",
    "adl-10-cam0-rgb.zip": "http://fenix.ur.edu.pl/~mkepski/ds/data/adl-10-cam0-rgb.zip",
    "adl-10-data.csv":"http://fenix.ur.edu.pl/~mkepski/ds/data/adl-10-data.csv"
}
# Function to download files (if needed)
def download_file(url, filename,folderpath):
    import requests
    try:
        print(f"Downloading {filename} from {url}...")
        response = requests.get(url)
        response.raise_for_status()
        file_path = os.path.join(folderpath, filename)
        with open(file_path, "wb") as file:
            file.write(response.content)
        print(f"Downloaded {file_path } successfully.")
    except requests.exceptions.RequestException as e:
        print(f"Error downloading {filename}: {e}")

# Function to unzip all files in a given zip file
import zipfile

# Function to unzip all files in a directory
def unzip_files(data_dir):
    if not os.path.exists(data_dir):
        os.makedirs(data_dir)

    for file_name in os.listdir(data_dir):
        if file_name.endswith(".zip"):
            #print(data_dir)
            with zipfile.ZipFile(os.path.join(data_dir, file_name), 'r') as zip_ref:
                extract_dir = os.path.join(data_dir, file_name[:-4])  # Remove .zip
                if not os.path.exists(extract_dir):
                    os.makedirs(extract_dir)
                zip_ref.extractall(extract_dir)
                print(f"Extracted {file_name} to {extract_dir}")


# Function to load CSV data into a DataFrame
def load_csv_data(csv_file):
    if not os.path.exists(csv_file):
        raise FileNotFoundError(f"CSV file not found: {csv_file}")
    return pd.read_csv(csv_file)

def load_frames_from_folder(folder, time_steps=30, height=64, width=64):

    print(f"Loading frames from folder: {folder}")
    frame_files = sorted(glob.glob(os.path.join(folder, "**","*.png")))  # Adjust for actual frame extension
    if not frame_files:
        return None

    frames = []
    for i in range(0, len(frame_files) - time_steps + 1, time_steps):
        sequence = []

        for j in range(time_steps):

            img = load_img(frame_files[i + j], target_size=(height, width))
            img_array = img_to_array(img) / 255.0
            sequence.append(img_array)
        frames.append(np.array(sequence))
    return np.array(frames)

# Main function to process the data from `data_all`
def process_data_from_data_all(data_all_dir):
    X_data = []
    y_data = []
    # Loop through all folders in `data_all`
    for subdir, dirs, files in os.walk(data_all_dir):

        for dir_name in dirs:

            folder_path = os.path.join(subdir, dir_name)
            csv_file_path = os.path.join(subdir, f"{dir_name[:7]}-data.csv" if dir_name.startswith("fall") else f"{dir_name[:6]}-data.csv")
            print(folder_path,csv_file_path)
            if os.path.exists(folder_path) and os.path.exists(csv_file_path):
                print(f"Processing {folder_path} with {csv_file_path}")

                # Load frames
                frames = load_frames_from_folder(folder_path, time_steps=30)

                # Load and align labels
                csv_data = load_csv_data(csv_file_path)
                if 'fall' in folder_path:
                    labels = 1  # 'fall' folders are labeled as 1 (fall)
                else:
                    labels = 0

                if frames is not None:
                    X_data.extend(frames)
                    y_data.extend([labels] * len(frames))

    # Ensure data is not empty
    if not X_data:
        raise ValueError("No data found. Check the file paths and data directory structure.")

    X_data = np.array(X_data)
    y_data = np.array(y_data)

    # Split into train and test sets
    X_train, X_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2, random_state=42)
    return X_train, X_test, y_train, y_test

# ##Start
data_all_dir = "/content/data_all"
# if not os.path.exists(data_all_dir):
#     os.makedirs(data_all_dir)
# ##Download the dataset - use only once
# for filename, url in urls.items():
#     if not os.path.exists(filename):
#         download_file(url, filename,data_all_dir)
#     if filename.endswith(".zip"):
#           # Path to the folder containing the zipped data
#         unzip_files(data_all_dir)
#     print(f"Processed {filename}")

#  ###moving csv file to data dir
# !mv *.csv /content/data_all/
###Process the data from `data_all`

X_train, X_test, y_train, y_test = process_data_from_data_all(data_all_dir)

print("Data processed:")
print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")
# Build and train CNN-LSTM model
from tensorflow.keras import layers, models

def create_cnn_lstm_model(input_shape):
    model = models.Sequential()

    # CNN layers for spatial feature extraction within a TimeDistributed wrapper
    model.add(layers.TimeDistributed(layers.Conv2D(32, (3, 3), activation='relu'), input_shape=input_shape))
    model.add(layers.TimeDistributed(layers.MaxPooling2D((2, 2))))
    model.add(layers.TimeDistributed(layers.Conv2D(64, (3, 3), activation='relu')))
    model.add(layers.TimeDistributed(layers.MaxPooling2D((2, 2))))
    model.add(layers.TimeDistributed(layers.Conv2D(128, (3, 3), activation='relu')))
    model.add(layers.TimeDistributed(layers.MaxPooling2D((2, 2))))

    # Reshape before LSTM
    model.add(layers.TimeDistributed(layers.Flatten()))
    model.add(layers.LSTM(64, return_sequences=False))

    # Fully connected layer for classification
    model.add(layers.Dense(1, activation='sigmoid'))  # Sigmoid for binary classification

    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    return model

# Model input shape (30 frames of 64x64 images with 3 channels)
input_shape = (30, 64, 64, 3)
model = create_cnn_lstm_model(input_shape)

# Train the model
model.summary()
model.fit(X_train, y_train, batch_size=16, epochs=50, validation_data=(X_test, y_test))

# Evaluate the model
test_loss, test_acc = model.evaluate(X_test, y_test)
print(f"Test accuracy: {test_acc * 100:.2f}%")
```
